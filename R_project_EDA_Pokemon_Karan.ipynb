3. Calculate Mode

```r
# Create mode function
mode <- function(x) {
  unique_x <- unique(x)
  tabulate_x <- tabulate(match(x, unique_x))
  unique_x[tabulate_x == max(tabulate_x)]
}
```

```r
mode(Grass_pokemon$speed)
```

```r
mode(Grass_pokemon$hp)
```

```r
mode(Grass_pokemon$attack)
```

```r
mode(Grass_pokemon$defense)
```

```r
mode(Grass_pokemon$sp_attack)
```

```r
mode(Grass_pokemon$sp_defense)
```
